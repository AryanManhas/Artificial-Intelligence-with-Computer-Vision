```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image_path = 'apple.jpg'
img = cv2.imread(image_path)

# Check if the image is loaded correctly
if img is None:
    print("Could not load image. Please check the path.")
    exit()

# 1. Convert to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 2. Convert to Grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 3. Create a white image (all pixels 255)
white_image = np.ones((300, 300, 3), dtype=np.uint8) * 255

# 4. Create a black image (all pixels 0)
black_image = np.zeros((300, 300, 3), dtype=np.uint8)

# 5. Draw shapes on a copy of original image
img_shapes = img.copy()

# Draw a red rectangle (x1, y1), (x2, y2)
cv2.rectangle(img_shapes, (50, 50), (200, 200), (0, 0, 255), 3)

# Draw a green circle (center), radius
cv2.circle(img_shapes, (50, 60), 50, (0, 255, 0), -1)  # -1 means filled

# Draw a blue line from (x1, y1) to (x2, y2)
cv2.line(img_shapes, (0, 0), (300, 300), (255, 0, 0), 2)

# Add text
cv2.putText(img_shapes, 'OpenCV!', (50, 270), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)

# Function to display images in the output section using matplotlib
def show_img_in_output(title, image, is_gray=False):
    plt.figure(figsize=(6, 6))
    if is_gray:
        plt.imshow(image, cmap='gray')  # Show grayscale image
    else:
        plt.imshow(image)  # Show RGB image
    plt.title(title)
    plt.axis("off")  # Hide axes
    plt.show()

# Show images in the output section
show_img_in_output("Original", img)
show_img_in_output("RGB", img_rgb)
show_img_in_output("Grayscale", img_gray, is_gray=True)
show_img_in_output("White Image", white_image)
show_img_in_output("Black Image", black_image)
show_img_in_output("Shapes Image", img_shapes)

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load an image
image_path = 'apple.jpg'
img = cv2.imread(image_path)

# Check if the image is loaded correctly
if img is None:
    print("Could not load image. Please check the path.")
    exit()

# 1. Convert to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 2. Convert to Grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 3. Create a white image (all pixels 255)
white_image = np.ones((300, 300, 3), dtype=np.uint8) * 255

# 4. Create a black image (all pixels 0)
black_image = np.zeros((300, 300, 3), dtype=np.uint8)

# 5. Draw shapes on a copy of original image
img_shapes = img.copy()
cv2.rectangle(img_shapes, (50, 50), (200, 200), (0, 0, 255), 3)
cv2.circle(img_shapes, (50, 60), 50, (0, 255, 0), -1)
cv2.line(img_shapes, (0, 0), (300, 300), (255, 0, 0), 2)
cv2.putText(img_shapes, 'OpenCV!', (50, 270), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)

# Convert BGR images to RGB for plotting
img_rgb_display = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_shapes_rgb = cv2.cvtColor(img_shapes, cv2.COLOR_BGR2RGB)

# Store image data in a DataFrame
data = {
    "Title": ["Original", "RGB", "Grayscale", "White", "Black", "Shapes"],
    "Image": [img_rgb_display, img_rgb, img_gray, white_image, black_image, img_shapes_rgb]
}

df = pd.DataFrame(data)

# Plotting using the DataFrame
plt.figure(figsize=(18, 4))

for idx, row in df.iterrows():
    plt.subplot(1, len(df), idx + 1)
    if len(row["Image"].shape) == 2:
        plt.imshow(row["Image"], cmap='gray')
    else:
        plt.imshow(row["Image"])
    plt.title(row["Title"], fontsize=10)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
