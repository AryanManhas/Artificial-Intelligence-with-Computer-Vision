```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Step 1: Load the CSV file
df = pd.read_csv(r"C:\Users\ACER\Downloads\data_normalization_larger.csv")  # Corrected file path

print("Original Data:\n", df)

# Step 2: Select the columns you want to normalize
columns_to_normalize = ['Age', 'Income']  # Ensure these columns exist

# Ensure numeric data & handle missing values
df[columns_to_normalize] = df[columns_to_normalize].apply(pd.to_numeric, errors='coerce')
df[columns_to_normalize] = df[columns_to_normalize].fillna(df[columns_to_normalize].mean())  # Fill NaNs with mean

# Step 3: Apply Min-Max Normalization
scaler = MinMaxScaler()
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

print("\nNormalized Data:\n", df)

# Step 4: Save the normalized data to a new CSV
df.to_csv(r"C:\Users\ACER\Downloads\normalized_dataset.csv", index=False)

print("\nNormalized data saved as 'normalized_dataset.csv'")
```
