```python
# Data Cleaning (replacing with median)
import pandas as pd

# Load the dataset

file_path = "TotalCase.csv"

df = pd.read_csv(file_path)

# Drop unnecessary index column if it exists

if 'Unnamed: 0' in df.columns:

  df.drop(columns=['Unnamed: 0'], inplace=True)

# Replace all 0 values with the median of the respective column

for col in df.select_dtypes(include=['number']).columns: # Select only numerical columns

  median_value = df[col].median() # Compute the median

  df[col] = df[col].replace(0, median_value) # Replace 0 with median

# Display the entire DataFrame in Jupyter Notebook

print("Data Cleaned by replacing with median!")

df # Display the updated dataset

```

```python
import pandas as pd

from scipy.stats import zscore

# Load the dataset

file_path = "TotalCase.csv"

df = pd.read_csv(file_path)

# Drop unnecessary index column if it exists

if 'Unnamed: 0' in df.columns:

  df.drop(columns=['Unnamed: 0'], inplace=True)

# Replace all 0 values with the median of the respective column

for col in df.select_dtypes(include=['number']).columns: # Select only numerical columns

  median_value = df[col].median() # Compute the median

  df[col] = df[col].replace(0, median_value) # Replace 0 with median

# Apply Z-score normalization to numerical columns

numeric_cols = df.select_dtypes(include=['number']).columns # Select numeric columns

df[numeric_cols] = df[numeric_cols].apply(zscore) # Apply Z-score normalization

# Display the updated DataFrame in Jupyter Notebook

print("Data Normalized!")

df # Show the transformed dataset

```
